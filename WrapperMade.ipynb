```python
import pickle
import joblib
from sentence_transformers import SentenceTransformer
import numpy as np

# === Load wrapper ===
with open("bert_xgb_wrapper.pkl", "rb") as f:
    wrapper = pickle.load(f)

# Extract info from wrapper
embed_model_name = wrapper["embed_model_name"]
xgb_model_path = wrapper["xgb_model"]
label_encoder_path = wrapper["label_encoder"]

# === Load model + label encoder ===
xgb_model = joblib.load(xgb_model_path)
label_encoder = joblib.load(label_encoder_path)

# === Load embedder ===
embedder = SentenceTransformer(embed_model_name)

# === Sample inputs ===
sample_texts = [
    "Government announces new healthcare reforms to benefit rural areas.",
    "Breaking: Celebrity admits aliens control the weather in shocking interview!"
]

# === Generate embeddings ===
embeddings = embedder.encode(sample_texts, convert_to_numpy=True, show_progress_bar=False)

# === Predict ===
preds = xgb_model.predict(embeddings)
labels = label_encoder.inverse_transform(preds)

for text, label in zip(sample_texts, labels):
    print(f"\n📰 {text}\n→ Predicted Label: {label}")
```

```python
from sentence_transformers import SentenceTransformer
import joblib, numpy as np

texts = [
  "Grid Operators Admit ...",    # paste Article1 full text
  "Breakthrough Antiviral ...",  # Article2
  "Advisory Council Report ...", # Article3
  "Heliocore Announces ...",     # Article4
  "Planetary Radar Team ..."     # Article5
]

wr = joblib.load("bert_xgb_wrapper.pkl")
clf = wr["clf"]; bert = wr.get("bert_name","all-mpnet-base-v2")
model = SentenceTransformer(bert)
embs = model.encode(texts, convert_to_numpy=True, show_progress_bar=False)
preds = clf.predict(embs)
probas = clf.predict_proba(embs) if hasattr(clf, "predict_proba") else None

for i,t in enumerate(texts):
    print(i+1, "pred:", preds[i], "proba:", (probas[i] if probas is not None else "no-proba"))
```
